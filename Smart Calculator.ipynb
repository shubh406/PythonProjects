```python
from collections import deque


def infix_to_postfix(arr):
    other = deque()
    postfix = []
    operators = {'^':5,'/':4,'*':3,'+':2,'-':2,'(':-1,')':-2}
    for ele in arr:
        if isinstance(ele,int) or ele == '(':
            other.append(ele)
        elif isinstance(ele,str) and (ele not in operators.keys()):
            other.append(int(ele))
        elif ele in '-*/+^':
            while other and (isinstance(other[-1],int) or operators[other[-1]]>=operators[ele]):
                postfix.append(other.pop())
            other.append(ele)
        else:
            while other and other[-1]!='(':
                postfix.append(other.pop())
            other.pop()
    while other:
        postfix.append(other.pop())
    return postfix
                
    
def typeof(characters):
    if characters.isalpha():
        return 'variable'
    if characters.isnumeric() or characters[1:].isnumeric() and characters[0] in '+-':
        return 'number'
    if characters in '+-*/^()':
        return 'operator'
    return ' ' if characters == ' ' else ''


def parse(expression):
    infix = []
    start = 0
    current_type = typeof(expression[start])
    for end in range(1, len(expression)):
        if current_type == 'operator' or typeof(expression[end]) != current_type:
            if expression[start] != ' ':
                infix.append(expression[start:end])
            start = end
            current_type = typeof(expression[start])
    infix.append(expression[start:])
    return infix
    


def check(s):
    if s.isalpha() is False:
        if s.isnumeric() is False:
            print("Invalid indetifier")
            return True
        elif s.isnumeric():
            print(int(s))
            return True
    else:
        if s in univ_dict.keys():
            print(univ_dict[s])
            return True
        else:
            print("Unknown variable")
            return True
    return False
    
def eval_postfix(arr):
    stack = []
    operators = {'^':5,'/':4,'*':3,'+':2,'-':1}
    for i in arr:
        if i not in operators.keys():
            stack.append(i)
        else:
            var1 = stack[-1]
            stack = stack[:-1]
            var2 = stack[-1]
            stack = stack[:-1]
            push = None
            var = operators[i]
            if var == 5:
                push = var1 ** var2
            elif var == 4:
                push = var2 // var1
            elif var == 3:
                push = var1 * var2
            elif var == 2:
                push = var1 + var2
            else:
                push = var2 - var1
            stack.append(push)
    return stack[0]
        


def validate(exp):
    prev=''
    operators = {'^':5,'/':4,'*':3,'+':2,'-':2}
    output = []
    brackets = 0
    flag = True
    for i in range(len(exp)):
        j = exp[i]
        if j!=' ':
            if j in operators.keys():
                if prev in operators.keys():
                    if j == prev and (j == '+' or j == '-'):
                        output = output[:-1]
                        output.append('+')
                        prev = '+'
                    elif j != prev and (j == '+' or j == '-') and (prev == '+' or prev == '-'):
                        output = output[:-1]
                        output.append('-')
                        prev = '-'
                    else:
                        flag = False
                        break
                else:
                    output.append(j)
                    prev = j
            else:
                if j == '(' or j == ')':
                    output.append(j)
                    prev = j
                elif j.isalpha():
                    if j in univ_dict.keys():
                        output.append(int(univ_dict[j]))
                        prev = int(univ_dict[j])
                    else:
                        return "Unknown variable", False
                elif j.isalpha() is False:
                    if j.isnumeric():
                        output.append(j)
                        prev = j
                    else:
                        return "Invalid identifier",False
                
            index = i - 1
            if j == '(':
                brackets += 1
            if j == ')':
                brackets -= 1
    if brackets == 0:
        return output,True
    else:
        return "Invalid expression", False

    
flag = True
univ_dict = {}
while flag:
    string = input().strip()
    if string == "/exit":
        print("Bye!")
        flag = False
    elif string == "/help":
        print("Add and Subtract the sequence of numbers")
    elif len(string) > 0 and string[0] == '/':
        print("Unknown command")
    elif "=" in string:
        index = string.index('=')
        string1 = ""
        string2 = ""
        for i in range(index-1,-1,-1):
            if string[i] != ' ':
                string1 = string[:i+1]
                break
        for i in range(index+1,len(string)):
            if string[i] != ' ':
                string2 = string[i:]
                break
        string1 = string1.strip()
        string2 = string2.strip()
        if string1.isalpha() is False:
            print("Invalid identifier")
        elif string2.isnumeric() or string2[0] in '+-' and string2[1:].isnumeric():
            univ_dict[string1] = int(string2)
        elif string2.isalpha() is False:
            print("Invalid assignment")
        else:
            value = univ_dict.get(string2,"Unknown variable")
            if isinstance(value,int):
                univ_dict[string1] = value
            else:
                print(value)        
    elif len(string)!=0:
        ans = parse(string.strip())
        result, bool_f = validate(ans)
        if bool_f:
            postfix_expr = infix_to_postfix(result)
            ans = eval_postfix(postfix_expr)
            print(ans)
        else:
            print(result)
```
