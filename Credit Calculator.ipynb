```python
import argparse
import math


def calculate_diff(principal, periods, interest, payment):
    if payment != 0:
        error_msg()
        return
    interest = interest / 1200
    over_payment = -1 * principal
    for i in range(1, periods+1):
        d = (principal / periods) + interest * (principal - ((principal * (i - 1)) / periods))
        print("Month {}: paid out {}".format(i, math.ceil(d)))
        over_payment += math.ceil(d)
    print()
    print("Overpayment = ", math.ceil(over_payment))


def calculate_req(principal, periods, interest, payment):
    if principal == 0:
        i = interest / 1200
        value = (i * math.pow(1 + i, periods)) / (math.pow(1 + i, periods) - 1)
        answer = math.ceil(payment / value)
        print("Your credit principal = ", answer, "!", sep="")
        overpayment = payment * periods - answer
        print("Overpayment = ", math.floor(overpayment))
    elif payment == 0:
        i = interest / 1200
        value = (i * math.pow(1 + i, periods)) / (math.pow(1 + i, periods) - 1)
        answer = math.ceil(value * principal)
        print("Your annuity payment = ", answer, "!", sep="")
        overpayment = answer * periods - principal
        print("Overpayment = ", math.floor(overpayment))
    elif periods == 0:
        i = interest / 1200
        base = 1 + i
        value = payment / (payment - (i * principal))
        ans = math.ceil(math.log(value, base))
        months = ans % 12
        years = ans // 12
        if months == 0:
            print("You need ", years, " years to repay this credit!", sep="")
        elif years == 0:
            print("You need ", months, " months to repay this credit!", sep="")
        else:
            print("You need ", years, " years and ", months, " months to repay this credit!", sep="")
        overpayment = ans * payment - principal
        print("Overpayment = ", math.floor(overpayment))
    else:
        error_msg()
        exit()


def error_msg():
    print("Incorrect parameters.")
    exit()


parser = argparse.ArgumentParser()
parser.add_argument("--type", type=str, default=None)
parser.add_argument("--principal", type=int, default=0)
parser.add_argument("--periods", type=int, default=0)
parser.add_argument("--payment", type=int, default=0)
parser.add_argument("--interest", type=float, default=0.0)
args = parser.parse_args()
if args.interest <= 0:
    error_msg()
elif args.type == "diff":
    calculate_diff(args.principal, args.periods, args.interest, args.payment)
elif args.type == "annuity":
    calculate_req(args.principal, args.periods, args.interest, args.payment)
else:
    error_msg()
```
