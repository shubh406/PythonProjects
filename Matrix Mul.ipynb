```python
def arr_input(x):
    if '.' in x:
        return float(x)
    return int(x)

def minor(arr,arr1):
    if len(arr) == 2:
        return arr[0][0] * arr[1][1] - arr[0][1]*arr[1][0]
    sign = 1
    for i in range(len(arr)):
        for j in range(len(arr[0])):
            arr_new = [[arr[x][y] for y in range(len(arr[0])) if y!= j ] for x in range(len(arr)) if x!= i]
            arr1[i][j] = det(arr_new) * (-1)**(i+j)


def det(arr):
    if len(arr) == 1 and len(arr[0]) == 1:
        return arr[0][0]
    if len(arr) == 2:
        return arr[0][0] * arr[1][1] - arr[0][1]*arr[1][0]
    ans1 = 0
    sign = 1
    for i in range(len(arr[0])):
        arr_new = [[arr[y][x] for x in range(len(arr[0])) if x!=i] for y in range(1,len(arr))]
        ans1 += arr[0][i]*det(arr_new)*sign
        sign = sign * -1 
    return ans1
    
def input_output():
    print("1. Add matrices")
    print("2. Multiply matrix by a constant")
    print("3. Multiply matrices")
    print("4. Transpose matrix")
    print("5. Calculate a determinant")
    print("6. Inverse matrix")
    print("0. Exit")
    
    choice = int(input("Your choice: > "))
    
    if choice == 1:
        add_matrix()
        input_output()
    if choice == 2:
        multiply()
        input_output()
    if choice == 3:
        mat_mul()
        input_output()
    if choice == 4:
        transpose()
        input_output()
    if choice == 5:
        determinant()
        input_output()
    if choice == 6:
        inverse()
        input_output()
    if choice == 0:
        return


def add_matrix():
    string = input("Enter size of first matrix: > ")
    row1, col1 = (int(x) for x in string.split())
    print("Enter first matrix:")
    arr1 = [[arr_input(x) for x in input().split()]for y in range(row1)]
    string = input("Enter size of second matrix: > ")
    row2, col2 = (int(x) for x in string.split())
    print("Enter second matrix:")
    arr2 = [[arr_input(x) for x in input().split()]for y in range(row2)]
    if row1 == row2 and col1 == col2:
        print("The result is:")
        arr3 = [[0 for i in range(col1)] for k in range(row1)]
        for i in range(row1):
            for j in range(col1):
                arr3[i][j] = arr1[i][j]+arr2[i][j]
        for i in arr3:
            print(*i)
    else:
        print("ERROR")
    return


def multiply():
    string = input("Enter size of matrix: > ")
    row1, col1 = (int(x) for x in string.split())
    print("Enter matrix:")
    arr1 = [[arr_input(x) for x in input().split()]for y in range(row1)]
    constant = arr_input((input("Enter constant: > ")))
    print("The result is:")
    for i in range(row1):
        for j in range(col1):
            print(arr1[i][j] * constant,end = " ")
        print()
    return
    

def mat_mul():
    string = input("Enter size of first matrix: > ")
    row1, col1 = (int(x) for x in string.split())
    print("Enter first matrix:")
    arr1 = [[arr_input(x) for x in input().split()]for y in range(row1)]
    string = input("Enter size of second matrix: > ")
    row2, col2 = (int(x) for x in string.split())
    print("Enter second matrix:")
    arr2 = [[arr_input(x) for x in input().split()]for y in range(row2)]
    if col1 == row2:
        print("The result is:")
        arr3 = [[0 for i in range(col2)] for j in range(row1)]
        for i in range(row1):
            for j in range(col2):
                for k in range(row2):
                    arr3[i][j] += arr1[i][k] * arr2[k][j]
        for i in arr3:
            print(*i)
    else:
        print("ERROR")
    return


def transpose():
    string = '''1. Main diagonal\n2. Side diagonal\n3. Vertical line\n4. Horizontal line'''
    print(string)
    choice = int(input("Your choice: > "))
    string = input("Enter matrix size: > ")
    row1, col1 = (int(x) for x in string.split())
    print("Enter matrix:")
    arr = [[arr_input(x) for x in input().split()]for y in range(row1)]
    print("The result is:")
    if choice == 1:
        for i in range(row1):
            for j in range(col1):
                if i < j:
                    arr[i][j],arr[j][i] = arr[j][i],arr[i][j]
    elif choice == 2:
        for i in range(row1):
            count = 1
            for j in range(col1):
                if i + j >= row1 :
                    arr[i][j],arr[i-count][j-count] = arr[i-count][j-count],arr[i][j]
                    count += 1
    elif choice == 3:
        for i in range(row1):
            for j in range(col1//2):
                j1 = col1 - 1 - j
                arr[i][j],arr[i][j1] = arr[i][j1],arr[i][j]
    elif choice == 4:
        for i in range(row1//2):
            for j in range(col1):
                i1 = row1 - 1 - i
                arr[i][j],arr[i1][j] = arr[i1][j],arr[i][j]
    for i in arr:
        print(*i)
    return


def determinant():
    string = input("Enter matrix size: > ")
    row1, col1 = (int(x) for x in string.split())
    print("Enter matrix:")
    arr = [[arr_input(x) for x in input().split()]for y in range(row1)]
    ans = det(arr)
    print("The result is:")
    print(ans)
    return
        
    
def inverse():
    string = input("Enter matrix size: > ")
    row1, col1 = (int(x) for x in string.split())
    print("Enter matrix:")
    arr = [[arr_input(x) for x in input().split()]for y in range(row1)]
    ans = det(arr)
    if ans == 0:
        print("This matrix doesn't have an inverse.")
        return
    else:
        arr1 = [[0 for i in range(col1)] for j in range(row1)]
        minor(arr,arr1)
        for i in range(row1):
            for j in range(col1):
                if i < j:
                    arr1[i][j],arr1[j][i] = arr1[j][i],arr1[i][j]
        for i in range(row1):
            for j in range(col1):
                arr1[i][j] /= ans
                if arr1[i][j] == 0.0 or arr1[i][j] == -0.0:
                    arr1[i][j] = 0
        for i in arr1:
            print(*i)
        return
        

input_output()
```
