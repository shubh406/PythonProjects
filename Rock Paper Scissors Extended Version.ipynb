```python
import random

global univ_dict
global total_options

univ_dict = {}
total_options= []

def get_result(computer_turn,user_turn,univ_dict):
    
    if computer_turn in univ_dict[user_turn]:
        return "Sorry, but computer chose " + str(computer_turn), 0
    elif user_turn == computer_turn:
        return "There is a draw (" + str(computer_turn) + ')' , 1
    else:
        return "Well Done. Computer chose " + str(computer_turn) + " and failed", 2
        

def computer_choice(total_options):
    choices = [i for i in range(len(total_options))]
    computer_option = random.choice(choices)
    return total_options[computer_option]

def make_new_dict(total_options):
    
    dump_list = []
    for i in range(len(total_options)):
        first_end_index = len(total_options)-i 
        dump_list[0:first_end_index] = total_options[i+1:]
        dump_list[first_end_index:] = total_options[0:i]
        index = len(dump_list) // 2
        key = total_options[i]
        univ_dict[key] = dump_list[0:index]
        #print(univ_dict[key])
        dump_list = []
    return univ_dict
        

def input_output():
    
    user_name = input("Enter your name: ")
    print("Hello, ",user_name)
    
    total_options,univ_dict = user_input()
    
    print("Okay,let's start")
    
    user_score = 0
    user_avail = False
    f = open('rating.txt','r+',encoding='utf-8')
    lst = f.readlines()
    for i in lst:
        name,score = i.split(" ")
        if name == user_name:
            user_score = int(score)
            user_avail = True
            break
    
    f.close()
    options_dict = ["!rating","!exit"]
    while True:
        user_turn = input()
        if (user_turn not in total_options) and (user_turn not in options_dict):
            print("Invalid input")
        elif user_turn == "!exit":
            print("Bye!")
            break
        elif user_turn == "!rating":
            print("Your rating:",user_score)
        else:
            computer_turn = computer_choice(total_options)
            ans,status = get_result(computer_turn,user_turn,univ_dict)
            if status == 1:
                user_score += 50
            elif status == 2:
                user_score += 100
            print(ans)
    
    
    new_content = user_name+" "+str(user_score)+"\n"
    new_file_content = ""
    if user_avail is True:
        f = open("rating.txt",'r',encoding = 'utf-8')
        lst = f.readlines()
        for i in lst:
            name,score = i.strip().split(" ")
            if name == user_name:
                new_i = i.strip()
                new_line = new_i.replace(score,str(user_score))
                new_file_content += new_line+"\n"
            else:
                new_i = i.strip()
                new_file_content += new_i+"\n"
                
        f.close()
        
        new_f = open('rating.txt','w',encoding = 'utf-8')
        new_f.write(new_file_content)
        new_f.close()
            
    else:
        f = open("rating.txt",'a+',encoding = 'utf-8')
        f.write(new_content)
        f.close()
        

def user_input():
    total_options = []
    univ_dict = {}
    total_options = input().strip().split(',')
    #print(len(total_options))
    
    if len(total_options) == 1:
        total_options = ["rock","paper","scissors"]
        univ_dict["scissors"] = "rock"
        univ_dict["rock"] = "paper"
        univ_dict["paper"] = "scissors"
    else:
        univ_dict = make_new_dict(total_options)
    return total_options, univ_dict
    
            
            
input_output()
```
